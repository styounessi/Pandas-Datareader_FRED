```python
import pandas_datareader as pdr
import plotly.express as px
import pandas as pd
from datetime import date
```

```python
start = datetime.datetime(2005, 1, 1)
end = date.today()

df = pdr.DataReader('PAYEMS', 'fred', start, end)

display(df)
```

```python
df = df*1000

display(df)
```

```python
fig = px.line(df, y='PAYEMS',
              title='Total Nonfarm Payrolls 2005 to Present',
              labels={'PAYEMS':'All Nonfarm Employees'})

fig.show(renderer="png")
```

```python
df = df['PAYEMS'].diff()

display(df)
```

```python
fig = px.bar(df, y='PAYEMS',
             title='Monthly Change in Payrolls, 2005 to Present',
             labels={'PAYEMS':'Monthly Payroll Change'})

fig.show(renderer="png")
```
