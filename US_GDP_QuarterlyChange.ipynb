```python
import pandas_datareader as pdr
import pandas as pd
import datetime
from datetime import date
import altair as alt
```

```python
start = datetime.datetime(1947, 1, 1)
end = date.today()

df = pdr.DataReader('A191RL1Q225SBEA', 'fred', start, end)
df.reset_index(level=0, inplace=True) 

display(df)
```

```python
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('DATE:T', axis=alt.Axis(format='', title='Date')),
    y=alt.Y('A191RL1Q225SBEA:Q', axis=alt.Axis(format='~s', title='% Change from Preceding Period')),
    color=alt.condition(
        alt.datum.A191RL1Q225SBEA > 0,
        alt.value('steelblue'),
        alt.value('red'))
).properties(
    title='Quarterly Change in Real Gross Domestic Product - United States',
    width=900,
    height=350)

chart.configure_title(
    fontSize=25,
    anchor='middle',
    color='black')
```
