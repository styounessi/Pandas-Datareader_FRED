```python
import pandas_datareader as pdr
import pandas as pd
import plotly.express as px
import datetime
from datetime import date
```

```python
start = datetime.datetime(2005, 1, 1)
end = date.today()

df = pdr.DataReader(['LNS13023653', 'LNS13026638'],
                    'fred', start, end)

df = df*1000

display(df)
```

```python
df = df.rename(columns={'LNS13023653':'Temporarily Laid Off',
                        'LNS13026638':'Permanently Laid Off'})

display(df)
```

```python
fig = px.line(df, y=['Temporarily Laid Off',
                     'Permanently Laid Off'],
              title='Permanent vs Temporary Layoffs, 2005 to Present',
              labels={'value':'Total',
                      'variable':'Status'})

fig.show(renderer="png")
```
