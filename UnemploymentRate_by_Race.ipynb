```python
import pandas_datareader as pdr
import pandas as pd
import plotly.express as px
import datetime
from datetime import date
```

```python
start = datetime.datetime(2000, 1, 1)
end = date.today()

df = pdr.DataReader(['LNS14000003', 'LNS14000006', 'LNS14000009','LNU04032183'],
                    'fred', start, end)

df = df.rename(columns={'LNS14000003':'White',
                        'LNS14000006':'Black',
                        'LNS14000009':'Hispanic or Latino',
                        'LNU04032183':'Asian'})
display(df)
```

```python
fig = px.line(df, y=['White',
                     'Black',
                     'Hispanic or Latino',
                     'Asian'],
              title='Unemployment Rate by Race, 2000 to Present',
              labels={'value':'Percent',
                      'variable':'Race'})

fig.show(renderer='png')
```

```python
df2 = df.tail(6)

display(df2)
```

```python
fig = px.bar(df2, y=['White',
                     'Black',
                     'Hispanic or Latino',
                     'Asian'],
             barmode='group',
             title='Recent Unemployment Rate by Race',
             text='value',
             labels={'value':'Percent',
                      'variable':'Race'})

fig.update_layout(xaxis_tickangle=-45)
fig.show(renderer='png', width=1300, height=500)
```
