```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime, timedelta
import plotly.express as px
```

```python
start = datetime.now() + timedelta(weeks=-52)
end = datetime.now()

df = pdr.DataReader('CPALTT01USM659N', 'fred', start, end)
df = df.rename(columns={'CPALTT01USM659N':'CPI Year over Year'}) 
df = df['CPI Year over Year'].round(decimals=3)

display(df)
```

```python
fig = px.bar(df, y='CPI Year over Year',
                 text='CPI Year over Year')

fig.update_traces(textposition='inside')
fig.update_layout(xaxis_tickangle=-45)

fig.show(renderer='png', width=1000, height=600)
```
