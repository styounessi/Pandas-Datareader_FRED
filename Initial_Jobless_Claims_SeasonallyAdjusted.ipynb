```python
import pandas_datareader as pdr
import pandas as pd
import plotly.express as px
import datetime
from datetime import date
```

```python
start = datetime.datetime(2019, 6, 1)
end = date.today()

df = pdr.DataReader(['ICSA', 'IC4WSA'],
                    'fred', start, end)

df = df.rename(columns={'ICSA':'Initial Jobless Claims',
                        'IC4WSA':'4 Week Moving Average'})

display(df)
```

```python
fig = px.line(df, y=['Initial Jobless Claims',
                     '4 Week Moving Average'],
              title='Initial Jobless Claims',
              labels={'value':'Total',
                      'variable':'Claims'})

fig.show(renderer="png")
```
